# OCR

```python
import os
from pdf2image import convert_from_path
from ultralytics import YOLO
import pdfplumber
from PIL import Image
import matplotlib.pyplot as plt
import cv2

# -------------------
# 1) ตั้งค่า
# -------------------
pdf_path = "/content/12_สายงานขายและปฏิบัติการลูกค้า 2.pdf"
model = YOLO("/content/detect head_mainpoint.pt")

# โฟลเดอร์เก็บภาพ
output_dir = "output_images"
os.makedirs(output_dir, exist_ok=True)

# -------------------
# 2) แปลง PDF → รูป
# -------------------
images = convert_from_path(pdf_path, dpi=200)  # dpi 150-300 กำลังดี

with pdfplumber.open(pdf_path) as pdf_file:
    pdf_width, pdf_height = pdf_file.pages[0].width, pdf_file.pages[0].height  # assume ทุกหน้าเหมือนกัน

final_images = []

for i, img in enumerate(images):
    # resize ให้ตรงกับ PDF
    img_resized = img.resize((int(pdf_width), int(pdf_height)), Image.LANCZOS)
    final_images.append(img_resized)

    # save image
    img_path = os.path.join(output_dir, f"page_{i+1}.png")
    img_resized.save(img_path, "PNG")
    print(f"Saved page {i+1} as {pdf_width}x{pdf_height} PNG")

# -------------------
# 3) Detect YOLO เก็บ bounding box
# -------------------
page_boxes = []

for i, img in enumerate(final_images):
    img_path = os.path.join(output_dir, f"page_{i+1}.png")

    results = model(img_path)

    # แสดงผล detect (optional)
    for r in results:
        im_array = r.plot()
        plt.imshow(cv2.cvtColor(im_array, cv2.COLOR_BGR2RGB))
        plt.axis("off")
        plt.show()

    boxes = []
    for result in results:
        for box in result.boxes.xyxy:
            boxes.append(tuple(box.tolist()))

    page_boxes.append({"page": i+1, "boxes": boxes})

print("Bounding boxes per page:")
for pb in page_boxes:
    print(pb)

# -------------------
# 4) ครอปข้อความจาก PDF ตาม bounding box
# -------------------
all_text = ""

with pdfplumber.open(pdf_path) as pdf:
    for pb in page_boxes:
        page_num = pb["page"]

        # ข้ามหน้า 1 และ 2
        if page_num in [1, 2]:
            continue

        page = pdf.pages[page_num - 1]

        # สมมติ boxes[0] = main, boxes[1] = head
        boxes = pb["boxes"]
        if len(boxes) < 2:
            continue  # ถ้าไม่มี head+main ก็ข้าม

        main_box = boxes[0]
        head_box = boxes[1]

        # pdfplumber ใช้ (x0, top, x1, bottom)
        x0, y0, x1, y1 = head_box

        padding_x = 3  # เพิ่ม/ลดแนวนอน 2 pt
        padding_y = 3  # เพิ่ม/ลดแนวตั้ง 2 pt
        x0 = max(0, x0 - padding_x)
        y0 = max(0, y0 - padding_y)
        x1 = min(960 , x1 + padding_x)
        y1 = min(540, y1 + padding_y)

        crop_box = (x0, y0, x1, y1)

        head_crop = page.within_bbox((crop_box))
        head_text = head_crop.extract_text()

        x0, y0, x1, y1 = main_box
        main_crop = page.within_bbox((x0, y0, 960.0, 540.0))
        main_text = main_crop.extract_text()

        all_text += f"\n--Page--{head_text}\n{main_text}"

# -------------------
# 5) บันทึกผลรวม
# -------------------
# with open("detected_text.txt", "w", encoding="utf-8") as f:
#     f.write(all_text)
```

```python
import re

def ocr_to_markdown(ocr_text, structure_file="auto_structure.md"):
    # กฎการ map heading - ใช้ --Page-- เป็นส่วนหนึ่งของ pattern
    rules = [
        (r'^--Page--สายงาน', "#"),
        (r'^--Page--สำนัก', "#"),
        (r'^--Page--กลุ่ม', "##"),
        (r'^--Page--ฝ่าย', "###"),
        (r'^--Page--ส่วน', "####"),
    ]

    # เตรียม OCR lines
    if isinstance(ocr_text, str):
        ocr_lines = [l.rstrip() for l in ocr_text.splitlines() if l.strip()]
    else:
        ocr_lines = [l.rstrip() for l in ocr_text if l.strip()]

    md_output = []
    structure_lines = []
    buffer_line = ""
    buffer_type = None  # None / "*" / "-"
    last_heading = None  # เก็บ heading ล่าสุด

    for i, line in enumerate(ocr_lines):
        line = line.strip()
        if not line:
            continue

        # 🔹 ตรวจสอบ heading ตาม rules ที่มี --Page--
        matched = False
        for pattern, hashes in rules:
            if re.match(pattern, line):
                # flush buffer ก่อนหน้า
                if buffer_line:
                    md_output.append(f"{buffer_type} {buffer_line}")
                    buffer_line, buffer_type = "", None

                # แยกชื่อหัวข้อ (เอา --Page-- ออก)
                heading_text = line.replace("--Page--", "").strip()

                # เคส (ต่อ) → ไม่เพิ่ม heading ใหม่
                if "(ต่อ)" not in heading_text:
                    clean_heading_text = heading_text.replace("(ต่อ)", "").strip()
                    heading = f"{hashes} {clean_heading_text}"
                    md_output.append(heading)
                    structure_lines.append(heading)
                    last_heading = heading

                matched = True
                break

        if matched:
            continue

        # 🔹 ตรวจสอบ numbered list ที่อาจไม่มีเนื้อหา
        # sub-numbered list เช่น 1.1, 2.3
        m_sub = re.match(r'^(\d+\.\d+)\s*(.*)', line)
        # main numbered list เช่น 1., 2.
        m_main = re.match(r'^(\d+)\.\s*(.*)', line)

        if m_sub:
            if buffer_line:
                md_output.append(f"{buffer_type} {buffer_line}")
                buffer_line, buffer_type = "", None
            content = m_sub.group(2).strip()
            if content:  # มีเนื้อหา
                buffer_line, buffer_type = content, "-"
            # ถ้าไม่มีเนื้อหา ให้ skip (รอเนื้อหาบรรทัดถัดไป)
            continue

        elif m_main:
            if buffer_line:
                md_output.append(f"{buffer_type} {buffer_line}")
                buffer_line, buffer_type = "", None
            content = m_main.group(2).strip()
            if content:  # มีเนื้อหา
                buffer_line, buffer_type = content, "*"
            # ถ้าไม่มีเนื้อหา ให้ skip (รอเนื้อหาบรรทัดถัดไป)
            continue

        # 🔹 เนื้อหาทั่วไป (รวมถึงเนื้อหาที่ตามหลังเลขที่ไม่มีเนื้อหา)
        if buffer_line:
            buffer_line += " " + line
        else:
            # ถ้าบรรทัดนี้ไม่ใช่ตัวเลข ให้เป็น bullet point
            if not re.match(r'^\d+\.?\s*$', line):  # ไม่ใช่แค่ตัวเลขเปล่าๆ
                buffer_line, buffer_type = line, "*"

    # flush buffer สุดท้าย
    if buffer_line:
        md_output.append(f"{buffer_type} {buffer_line}")

    # 🔹 เขียน structure file
    with open(structure_file, "w", encoding="utf-8") as f:
        f.write("\n".join(structure_lines))

    return "\n".join(md_output)

# ----------------------
# ใช้งาน
# final_result = open("ocr_result.txt","r",encoding="utf-8").read()

# output_md = ocr_to_markdown(final_result, "3_สายงานการเงิน_โครงสร้างสายงาน.md")
# with open("3_สายงานการเงิน_หน้าที่รับผิดชอบสายงาน.md", "w", encoding="utf-8") as f:
#     f.write(output_md)

output_md = ocr_to_markdown(final_result, "/content/Md โครงสร้าง/12_สายงานขายและปฏิบัติการลูกค้า 2_โครงสร้างสายงาน.md")
with open("/content/หน้าที่รับผิดชอบสายงาน/12_สายงานขายและปฏิบัติการลูกค้า 2_หน้าที่รับผิดชอบสายงาน.md", "w", encoding="utf-8") as f:
    f.write(output_md)

print(output_md[:1000])
```

```python
import re
from pythainlp.tokenize import sent_tokenize

def process_md_for_rag(md_text, source="doc.md", max_chunk_size=500):
    """
    ประมวลผลไฟล์ .md ที่ได้จาก OCR -> ทำ chunk -> เก็บ hierarchy
    """
    # Regex จับหัวข้อ ##, ###, ####
    header_pattern = r'^(#{1,4})\s+(.*)'

    lines = md_text.splitlines()
    hierarchy_stack = []  # เก็บลำดับหัวข้อ
    chunks = []
    chunk_counter = 1
    current_chunk = ""

    def flush_chunk():
        nonlocal chunk_counter, current_chunk
        if current_chunk.strip():
            chunks.append({
                "source": source,
                "chunk_id": chunk_counter,
                "hierarchy": hierarchy_stack.copy(),
                "text": current_chunk.strip()
            })
            chunk_counter += 1
            current_chunk = ""

    for line in lines:
        header_match = re.match(header_pattern, line.strip())
        if header_match:
            # flush chunk ก่อนเปลี่ยนหัวข้อ
            flush_chunk()

            level = len(header_match.group(1))  # # = 1, ## = 2, ...
            header_text = header_match.group(2).strip()

            # ปรับ hierarchy stack ตาม level
            hierarchy_stack = hierarchy_stack[:level-1]
            hierarchy_stack.append(f"{'#'*level} {header_text}")
        else:
            # สะสมเนื้อหา
            sentence_list = sent_tokenize(line.strip(), engine="whitespace")
            for sentence in sentence_list:
                if len(current_chunk) + len(sentence) <= max_chunk_size:
                    current_chunk += sentence + " "
                else:
                    flush_chunk()
                    current_chunk = sentence + " "

    # flush chunk สุดท้าย
    flush_chunk()

    return chunks

# if __name__ == "__main__":
#     processed_chunks = process_md_for_rag(
#     final_result,
#     source="4_สายงานยุทธศาสตร์องค์กร_หน้าที่รับผิดชอบสายงาน.md",
#     max_chunk_size=1500
# )


#     import json
#     print(json.dumps(processed_chunks, ensure_ascii=False, indent=2))
```

```python
from transformers import AutoTokenizer, AutoModel
import torch
import numpy as np
from tqdm import tqdm

# โหลด tokenizer และ model
tokenizer = AutoTokenizer.from_pretrained("airesearch/wangchanberta-base-att-spm-uncased")
model = AutoModel.from_pretrained("airesearch/wangchanberta-base-att-spm-uncased")

MAX_TOKENS = 512  # token limit ของ WangchanBERT

def get_thai_embedding(text):
    try:
        inputs = tokenizer(text, return_tensors="pt", truncation=True, max_length=MAX_TOKENS)
        if inputs["input_ids"].shape[1] == 0:
            print(f"⚠️ No tokens found in text: {repr(text[:50])}")
            return None
        with torch.no_grad():
            outputs = model(**inputs)
        return outputs.last_hidden_state.mean(dim=1).squeeze().numpy()
    except Exception as e:
        print(f"🚨 Error embedding text: {repr(text[:50])} -> {e}")
        return None

def split_text_by_tokens(text, max_tokens=MAX_TOKENS):
    """
    แบ่งข้อความเป็นชิ้นโดยใช้ tokenizer เพื่อไม่เกิน max_tokens
    """
    words = text.split()
    chunks = []
    current_chunk = []

    for word in words:
        current_chunk.append(word)
        token_count = len(tokenizer(" ".join(current_chunk))["input_ids"])
        if token_count > max_tokens:
            # เอาก่อนหน้าเป็น chunk
            current_chunk.pop()  # เอาคำปัจจุบันออก
            chunks.append(" ".join(current_chunk))
            current_chunk = [word]  # เริ่ม chunk ใหม่
    if current_chunk:
        chunks.append(" ".join(current_chunk))
    return chunks

def process_chunks_to_embeddings(all_chunks):
    embeddings_data = []

    for chunk in tqdm(all_chunks, desc="Embedding chunks"):
        text = chunk["text"].strip()
        if not text:
            continue

        # แบ่งตาม token
        text_parts = split_text_by_tokens(text, max_tokens=MAX_TOKENS)
        part_embeddings = []
        for part in text_parts:
            emb = get_thai_embedding(part)
            if emb is not None:
                part_embeddings.append(emb)

        if part_embeddings:
            combined_embedding = np.mean(part_embeddings, axis=0)
            embeddings_data.append({
                "source": chunk.get("source", "unknown"),
                "chunk_id": chunk.get("chunk_id"),
                "hierarchy": chunk.get("hierarchy", []),
                "text": text,
                "embedding": combined_embedding
            })

    return embeddings_data
```

```python
import faiss
import numpy as np

# เตรียม numpy array ของ vectors เท่านั้น (สมมติ embedding dimension เท่ากันทั้งหมด)
vectors = np.array([item['embedding'] for item in thai_embeddings]).astype('float32')

# สร้าง FAISS index แบบ Flat (simple, exact search)
dimension = vectors.shape[1]  # ขนาด embedding dimension
index = faiss.IndexFlatL2(dimension) # ใช้ระยะห่างแบบ L2 (Euclidean)

# เพิ่ม vectors เข้า index
index.add(vectors)

# ทดสอบค้นหา
query_vector = vectors[0].reshape(1, -1)  # สมมติใช้ vector ตัวแรกค้นหา
k = 10  # จำนวนผลลัพธ์ที่ต้องการ

distances, indices = index.search(query_vector, k)
print("Nearest neighbors indices:", indices)
print("Distances:", distances)

# ดึง metadata ที่เกี่ยวข้อง (page, chunk_id) จาก indices
for idx in indices[0]:
    meta = thai_embeddings[idx]
    print(f"Source: {meta['source']}, Chunk ID: {meta['chunk_id']}, Hierarchy: {meta['hierarchy']}")
```

แปลงคำถามเป็น embedding ด้วยโมเดลเดียวกันหรือใกล้เคียง

ดึงข้อมูลที่ใกล้เคียงจาก vector store มาเป็น context (retrieval step)

```python
# เรียกโมเดล embeddingอีกครั้ง
from sentence_transformers import SentenceTransformer
import numpy as np

# โหลดโมเดล embedding ภาษาไทย
embed_model = SentenceTransformer("airesearch/wangchanberta-base-att-spm-uncased")

def get_thai_embedding(text: str) -> np.ndarray:
    embedding = embed_model.encode([text])[0]
    return np.array(embedding).astype('float32')
```

```python
def ask_rag(query_text: str, index, thai_embeddings, k=5):
    # 1. Embed คำถาม
    query_embedding = get_thai_embedding(query_text).reshape(1, -1)

    # 2. ดึง context จาก FAISS
    distances, indices = index.search(query_embedding, k=k)
    retrieved_contexts = [thai_embeddings[i]['text'] for i in indices[0]]

    # 3. รวมเป็น prompt
    context = "\n\n".join(retrieved_contexts)
    prompt = f"""ข้อมูลที่เกี่ยวข้อง:
{context}

คำถาม: {query_text}
กรุณาตอบคำถามจากข้อมูลข้างต้นอย่างชัดเจนและกระชับ"""

    return prompt
```

```python
import openai
import requests
import re

# vLLM
openai.api_base = "http://1.179.140.78:8000/v1"
openai.api_key = "nta100_test_00"

def generate_answer_with_vllm(prompt: str):
    # ประมาณการขนาด
    input_length = len(prompt)

    if input_length < 1000:      # ข้อความสั้น
        max_tokens = 800
    elif input_length < 3000:    # ข้อความปานกลาง
        max_tokens = 1500
    elif input_length < 6000:    # ข้อความยาว
        max_tokens = 2500
    else:                        # ข้อความยาวมาก
        max_tokens = 4000

    response = openai.ChatCompletion.create(
        model="Qwen/Qwen3-14B",
        messages=[
            {"role": "system", "content": "คุณคือตัวช่วยตอบคำถามจากข้อมูล"},
            {"role": "user", "content": prompt}
        ],
        max_tokens=max_tokens,
        temperature=0.3,
        enable_thinking=False
    )

    raw_output = response['choices'][0]['message']['content']
    answer = re.sub(r"<think>.*?</think>", "", raw_output, flags=re.DOTALL | re.IGNORECASE).strip() # clean_response
    return answer
```
